## Basic Visualization using Matplotlib

This notebook is to introduce you to the basics of visualization, using a simple data set to show different plot types

```python
# import needed libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating a pandas series
d = pd.Series(np.random.randn(10), index=pd.date_range('1/1/2019', periods=10, freq = 'M'))
```

Mind you that when you run this code yourself, you'll most likely not get same values
because of the random method used here.

```python
d.head(5)
```

```python
# Line plot
# Line plot is a grah used yo show frequency of data along a number line
d.plot()
```

```python
# Bar graph
# data values are proportional to the lenght of the bar
d.plot(kind = 'bar')
```

```python
# Histogram
# Histogram shows the frequency distribution of a data set
d.plot(kind = 'hist')
```

```python
'''Box plot is a standardized way of displaying the distribution of data.
   from the figure below it shows the maximum value to be 1.5,
   third quartile to be around 1.0, median = 0.0, 1st quartile to be around 0.5
   and minimum = -1.5'''
d.plot(kind = 'box')
```
