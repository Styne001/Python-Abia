## Basic Visualization using Matplotlib

This notebook is to introduce you to the basics of visualization, using a simple data set to show different plot types

```python
# import needed libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating a pandas series
d = pd.Series([2, 5, 4, 2, 3, 5, 2, 4, 3, 3], index=pd.date_range('1/1/2019', periods=10, freq = 'M'))
```

```python
d.head(5)
```

```python
# Line plot
# Line plot is a grah used yo show frequency of data along a number line
d.plot()
```

```python
# Bar graph
# data values are proportional to the lenght of the bar
d.plot(kind = 'bar')
```

```python
# Histogram
# Histogram shows the frequency distribution of a data set
d.plot(kind = 'hist')
```

```python
'''Box plot is a standardized way of displaying the distribution of data.
   from the figure below it shows the maximum value to be 5.0,
   third quartile to be around 4.0, median = 3.0, 1st quartile to be around 2.25
   and minimum = 2.0'''
d.plot(kind = 'box')
```
